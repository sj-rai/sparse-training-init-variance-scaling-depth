```python
results_small = {
    "Gaussian":   [98, 90.0, 20.1],
    "t_df2":      [97.8, 82.0, 30.3],
    "t_df5":      [96.0, 77.2, 25.5],
    "t_df10":     [96.1, 84.0, 30.8],
}

results_deep = {
    "Gaussian":   [98.5, 97.1, 72.2],
    "t_df2":      [98.2, 97.0, 71.0],
    "t_df5":      [98.4, 97.0, 70.0],
    "t_df10":     [98.3, 97.1, 78.3],
}

sparsity_levels = [90, 98, 99]
```

```python
import matplotlib.pyplot as plt

plt.figure()

for init_name, accs in results_small.items():
    plt.plot(sparsity_levels, accs, marker='o', label=init_name)

plt.xlabel("Sparsity (%)")
plt.ylabel("Average Test Accuracy (%)")
plt.title("Small CNN: Accuracy vs Sparsity")
plt.xticks(sparsity_levels)
plt.ylim(0, 100) 
plt.legend()
plt.savefig("small_cnn_accuracy.pdf")
plt.show()
```

```python
plt.figure()

for init_name, accs in results_deep.items():
    plt.plot(sparsity_levels, accs, marker='o', label=init_name)

plt.xlabel("Sparsity (%)")
plt.ylabel("Average Test Accuracy (%)")
plt.title("Deep CNN: Accuracy vs Sparsity")
plt.xticks(sparsity_levels)
plt.ylim(0, 100) 
plt.legend()
plt.savefig("deep_cnn_accuracy.pdf")
plt.show()
```
